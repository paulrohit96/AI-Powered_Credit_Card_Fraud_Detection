```python
import streamlit as st
import pandas as pd

# Load the dataset (replace with your actual dataset)
data = pd.read_csv("/content/fraud_cases_with_reasons.csv")

df = pd.DataFrame(data)

# Filter only the relevant columns for the case manager
relevant_columns = [
    "Customer_ID", "Transaction_Date", "Transaction_Amount", "Merchant_Name",
    "Merchant_Category", "Fraud_Flag", "Transaction_Hour", "Transaction_Day",
    "Transaction_Month", "Transaction_Frequency", "Location_Mismatch",
    "Fraud_Prediction", "Real_World_Reason", "Model_Reason"
]
df = df[relevant_columns]

# Streamlit App
st.title("Fraud Case Manager")
st.write("This is a case management tool for fraud investigators to review flagged transactions.")

# Display the filtered dataset
st.write("### Fraud Cases Overview")
st.dataframe(df)

# Allow the investigator to select a specific case for detailed review
case_id = st.selectbox("Select a Case ID to Investigate", df["Customer_ID"].unique())

# Display detailed information for the selected case
selected_case = df[df["Customer_ID"] == case_id].iloc[0]

st.write("### Case Details")
st.write(f"**Customer ID:** {selected_case['Customer_ID']}")
st.write(f"**Transaction Date:** {selected_case['Transaction_Date']}")
st.write(f"**Transaction Amount:** ${selected_case['Transaction_Amount']}")
st.write(f"**Merchant Name:** {selected_case['Merchant_Name']}")
st.write(f"**Merchant Category:** {selected_case['Merchant_Category']}")
st.write(f"**Fraud Flag:** {'Fraudulent' if selected_case['Fraud_Flag'] == 1 else 'Non-Fraudulent'}")
st.write(f"**Transaction Hour:** {selected_case['Transaction_Hour']}")
st.write(f"**Transaction Day:** {selected_case['Transaction_Day']}")
st.write(f"**Transaction Month:** {selected_case['Transaction_Month']}")
st.write(f"**Transaction Frequency:** {selected_case['Transaction_Frequency']}")
st.write(f"**Location Mismatch:** {'Yes' if selected_case['Location_Mismatch'] else 'No'}")
st.write(f"**Fraud Prediction:** {'Fraudulent' if selected_case['Fraud_Prediction'] == 1 else 'Non-Fraudulent'}")
st.write(f"**Real World Reason:** {selected_case['Real_World_Reason']}")
st.write(f"**Model Reason:** {selected_case['Model_Reason']}")

# Add a button to mark the case as resolved
if st.button("Mark Case as Resolved"):
    st.success(f"Case {case_id} has been marked as resolved.")
```
